## Generating Sample Data

```python
import numpy as np
import random
from matplotlib import pyplot as plt

data = np.random.normal(0, 20, 1000) 

# fixed bin size
bins = np.arange(-100, 100, 5) # fixed bin size

plt.xlim([min(data)-5, max(data)+5])

plt.hist(data)
plt.show()
```

## Deafault Grid

```python
plt.hist(data)
plt.grid()
plt.show()
```

## Vertical & Horizontal Grids

### Vertical Grid

```python
plt.hist(data)
ax = plt.gca()
ax.xaxis.grid(True)

plt.show()
```

### Horizontal Grid

```python
plt.hist(data)
ax = plt.gca()
ax.yaxis.grid(True)

plt.show()
```

## Chaging the Tick color and Frequency


```python
from matplotlib import rcParams

rcParams['grid.linestyle'] = '-'
rcParams['grid.color'] = 'blue'
rcParams['grid.linewidth'] = 0.2

plt.grid()
plt.hist(data)

plt.show()
```

## Formatting Legend

```python
x = np.arange(10)

for i in range(1, 4):
    plt.plot(x, i * x**2, label='Group %d' % i)

plt.legend(loc='best')
plt.show()
```

## Legend outside the box

```python
fig = plt.figure()
ax = plt.subplot(111)

x = np.arange(10)

for i in range(1, 4):
    ax.plot(x, i * x**2, label='Group %d' % i)

ax.legend(loc='upper center', 
          bbox_to_anchor=(0.5,  # horizontal
                          1.15),# vertical 
          ncol=3, fancybox=True)
plt.show()
```

```python
fig = plt.figure()
ax = plt.subplot(111)

x = np.arange(10)

for i in range(1, 4):
    ax.plot(x, i * x**2, label='Group %d' % i)

ax.legend(loc='upper center', 
          bbox_to_anchor=(1.15, 1.02),
          ncol=1, fancybox=True)
plt.show()
```

## Change in line style changes the Legend

```python
from itertools import cycle
x = np.arange(10)

colors = ['blue', 'red', 'green']
color_gen = cycle(colors)

for i in range(1, 4):
    plt.plot(x, i * x**2, label='Group %d' % i, marker='o')

plt.legend(loc='upper left', numpoints=1)
plt.show()
```

```python
from itertools import cycle
x = np.arange(10)

colors = ['blue', 'red', 'green']
color_gen = cycle(colors)

for i in range(1, 4):
    plt.scatter(x, i * x**2, label='Group %d' % i, color=next(color_gen))

plt.legend(loc='upper left', scatterpoints=1)
plt.show()
```
